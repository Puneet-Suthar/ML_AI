function to show image using opencv

```python
def image_show(image_path):
    '''function to read and show image from its path'''
    img = cv2.imread(image_path)
    cv2.imshow(image_path.split(".")[0],img)
    key = cv2.waitKey(0)
    if key == ord("q"):
        cv2.destroyAllWindows()
```

```python
image_show()
```

```python
image_show("robot-upflairs.png")
```
