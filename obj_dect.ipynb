# Libraries 

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.efficientnet import EfficientNetB0, preprocess_input, decode_predictions
import matplotlib.pyplot as plt
```

# loading image 

```python
image_name = 'hat.jpg'
img = cv2.imread(image_name)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

# Image tranformation for efficientNet

```python
img_resized = cv2.resize(img_rgb, (224,224))
img_processed = preprocess_input(img_resized)
img_ready = np.expand_dims(img_processed, axis = 0)
# plt.imshow(img_ready)
```

# Importing Model

```python
model = EfficientNetB0(weights = 'imagenet')
```

# Predicitons

```python
predictions = model.predict(img_ready)
dec_predictions = decode_predictions(predictions, top=5)[0]

print("Prediciton for the image:-")

for i,j in enumerate(dec_predictions):
    print(i, j[1], j[2])

plt.imshow(img_rgb)
plt.title(f'Pred:- {dec_predictions[0][1]} ({dec_predictions[0][2]:.2f})')
```
